```python
import matplotlib.pyplot as plt
import numpy as np

def YearlyPlots(T,ttl='',yl='',xl='',y=None,x=None,size=(10,7), c=None):
    yearday=[i for i in range(1,366)]
    fig=figure(1,figsize=size,dpi=300)
    if shape(T)[0] != 365:
        raise ValueError("First dimension of T should be 365. Shape(T)="+str(shape(T)))
    if c is not None:
        plot_date(yearday,T, '-',color=c);
        plt.gca().xaxis.set_major_formatter( DateFormatter('%b') )
          
    else:
        plot_date(yearday,T, '-', );
    # rotate and align the tick labels so they look better
        fig.autofmt_xdate()
        
    ylabel(yl)
    xlabel(xl)
    if y is not None:
        ylim(y)
    if x is not None:
        xlim(x)
    grid()
    title(ttl)
```

```python
figure(figsize=(15,30))
offset=1
for meas in STAT.keys():
    subplot(6,3,offset)
    offset+=1
    ## Your code for Histogram
    S=STAT[meas]
    low= S['low100']
    high= S['high100']
    X= S['SortedVals']
    Y=[]
    for i in xrange(0,X.shape[0]):
        if X[i]>=low and X[i]<=high:
            Y.append(X[i])
    restricted_array=array(Y)
    plt.hist(restricted_array,bins=100)
    grid()
    title(meas+' restricted histogram')
    
    subplot(6,3,offset)
    offset+=1
    ## Your code for mean and mean +- std
    YearlyPlots(S['Mean']+np.sqrt(S['Var']),ttl=meas+' mean +-std',yl='',xl='',y=None,x=None,size=(10,7), c='b')
    YearlyPlots(S['Mean'],ttl=meas+' mean +-std',yl='',xl='',y=None,x=None,size=(10,7), c='r')
    YearlyPlots(S['Mean']-np.sqrt(S['Var']),ttl=meas+' mean +-std',yl='',xl='',y=None,x=None,size=(10,7), c='b')    
    
    subplot(6,3,offset)
    offset+=1
    ## Your code for number of measurements
    YearlyPlots(S['NE']-S['std'],ttl=meas+' counts',yl='',xl='',y=None,x=None,size=(10,7), c='k')
```
